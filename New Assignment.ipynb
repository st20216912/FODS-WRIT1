```python
# Import libraries required
import pandas as pd
import numpy as np
from sklearn.impute import SimpleImputer
# Path saved to local work computer
path = "C:\\BSCADS\\FODS_data\\"
```

```python
# Import Ages csv into dataframe
age_df = pd.read_csv (path + "ages.csv")

# Replace 'unnamed' coloumn with 'Ages' and set to index
age_df.rename(columns = {'Unnamed: 0':'Ages'}, inplace = True)
age_df = age_df.set_index('Ages')
age_df
```

```python
# Import Road Class csv into dataframe
roadclass_df = pd.read_csv (path + "road_class.csv")

# Replace 'unnamed' coloumn with 'Class' and set to index
roadclass_df.rename(columns = {'Unnamed: 0' : 'Class'}, inplace = True)
roadclass_df = roadclass_df.set_index('Class')
roadclass_df
```

```python
# Import Days of Week csv into dataframe
days_df = pd.read_csv (path + "day_of_week.csv")

# Replace 'unnamed' coloumn with 'Days' and set to index
days_df.rename(columns = {'Unnamed: 0' : 'Days'}, inplace = True)
days_df = days_df.set_index('Days')
days_df
```

```python
# Import Gender csv into dataframe
gender_df = pd.read_csv (path + "gender.csv")

# Replace 'unnamed' coloumn with 'Gender' and set to index
gender_df.rename(columns = {'Unnamed: 0' : 'Gender'}, inplace = True)
gender_df = gender_df.set_index('Gender')
gender_df
```

```python
# Import Hit and Run csv into dataframe
hitrun_df = pd.read_csv (path + "hit_and_run.csv")

# Replace 'unnamed' coloumn with 'HitRun' and set to index
hitrun_df.rename(columns = {'Unnamed: 0' : 'HitRun'}, inplace = True)
hitrun_df = hitrun_df.set_index('HitRun')
hitrun_df
```

```python
# Import Junctions csv into dataframe
junctions_df = pd.read_csv (path + "junctions.csv")

# Replace 'unnamed' coloumn with 'Junctions' and set to index
junctions_df.rename(columns = {'Unnamed: 0' : 'Junctions'}, inplace = True)
junctions_df = junctions_df.set_index('Junctions')
junctions_df
```

```python
# Import Manoeuvres csv into dataframe
manoeuvres_df = pd.read_csv (path + "manoeuvres.csv")

# Replace 'unnamed' coloumn with 'Manoeuvres' and set to index
manoeuvres_df.rename(columns = {'Unnamed: 0' : 'Manoeuvres'}, inplace = True)
manoeuvres_df = manoeuvres_df.set_index('Manoeuvres')
manoeuvres_df
```

```python
# Import Road Type csv into dataframe
roadtype_df = pd.read_csv (path + "road_type.csv")

# Replace 'unnamed' coloumn with 'Road Type' and set to index
roadtype_df.rename(columns = {'Unnamed: 0' : 'Road Type'}, inplace = True)
roadtype_df = roadtype_df.set_index('Road Type')
roadtype_df
```

```python
# Import Speed Limit csv into dataframe
speedlimit_df = pd.read_csv (path + "speed_limit.csv")

# Replace 'unnamed' coloumn with 'Speed Limit' and set to index
speedlimit_df.rename(columns = {'Unnamed: 0' : 'Speed Limit'}, inplace = True)
speedlimit_df = speedlimit_df.set_index('Speed Limit')
speedlimit_df
```

```python
# Import Time of Day csv into dataframe
time_df = pd.read_csv (path + "time_of_day.csv")

# Replace 'unnamed' coloumn with 'Time' and set to index
time_df.rename(columns = {'Unnamed: 0' : 'Time'}, inplace = True)
time_df = time_df.set_index('Time')
time_df
```

```python
# Import Vehicle Type csv into dataframe
vehicle_df = pd.read_csv (path + "vehicle_type.csv")

# Replace 'unnamed' coloumn with 'Vehicle' and set to index
vehicle_df.rename(columns = {'Unnamed: 0' : 'Vehicle'}, inplace = True)
vehicle_df = vehicle_df.set_index('Vehicle')
vehicle_df
```

```python
# Import Weather Conditions csv into dataframe
weather_df = pd.read_csv (path + "weather_conditions.csv")

# Replace 'unnamed' coloumn with 'Weather' and set to index
weather_df.rename(columns = {'Unnamed: 0' : 'Weather'}, inplace = True)
weather_df = weather_df.set_index('Weather')
weather_df
```

```python
df = pd.concat([age_df,roadclass_df,days_df,gender_df,hitrun_df,junctions_df,manoeuvres_df,roadtype_df,speedlimit_df,time_df,vehicle_df,weather_df])
df.replace('.', np.nan, inplace = True)
df = df.T

df.isnull().sum()

titles=list(df)

years=['2015','2016','2017','2018','2019','2020']

x=df.iloc[:,0:81]
imputer=SimpleImputer(missing_values=np.nan,strategy='mean')
imputer=imputer.fit_transform(x)
df=pd.DataFrame(imputer)
df.index = years

df.columns = titles
df=df.astype(int)

df
```
